```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filepath = '/home/kvu/ddrl/benchmarks/openai-gyms/Acrobot-v1/result-seed-2021-2021_07_22.pickle'
```

```python
res = pd.read_pickle(filepath)
```

```python
scores = res['scores']
timepoints = res['timepoints']
means = res['means']
```

```python
n_episodes = len(means)
fig, ax = plt.subplots()
ax.plot(timepoints, means)
ax.set_xlabel('Training time (seconds)')
ax.set_ylabel('Mean score of 100 consecutive episodes')
ax.set_title('Acrobot-v1')
```

```python
fig.savefig('result.png')
```
