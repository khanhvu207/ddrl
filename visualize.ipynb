```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pylab import rcParams
rcParams['figure.figsize'] = 15, 10

```

```python
def extract_training_data(filepath):
    res = pd.read_pickle(filepath)
    scores = res['scores']
    timepoints = res['timepoints']
    means = res['means']
    return timepoints, means
```

```python
fig, ax = plt.subplots(2, 2)

fig.add_subplot(111, frameon=False)
fig.tight_layout()
plt.tick_params(labelcolor='none', top=False, bottom=False, left=False, right=False)
plt.grid(False)
plt.ylabel("Mean score of 100 consecutive episodes", fontsize=14)
plt.xlabel("Training time (seconds)", fontsize=14)

ax[0, 0].plot(*extract_training_data('/home/kvu/ddrl/benchmarks/openai-gyms/LunarLander-v2/result-seed-2021-2021_07_24_01_45_51.pickle'))
ax[0, 1].plot(*extract_training_data('/home/kvu/ddrl/benchmarks/openai-gyms/CartPole-v1/result-seed-2021-2021_07_24_02_14_29.pickle'))
ax[1, 0].plot(*extract_training_data('/home/kvu/ddrl/benchmarks/openai-gyms/Acrobot-v1/result-seed-2021-2021_07_24_02_27_21.pickle'))
ax[1, 1].plot(*extract_training_data('/home/kvu/ddrl/benchmarks/openai-gyms/MountainCar-v0/result-seed-2021-2021_07_25_20_18_32.pickle'))

ax[0, 0].set_title('LunarLander-v2')
ax[0, 1].set_title('CartPole-v1')
ax[1, 0].set_title('Acrobot-v1')
ax[1, 1].set_title('MountainCar-v0')
```

```python

fig, ax = plt.subplots()
ax.plot(timepoints, means)
ax.set_xlabel('Training time (seconds)')
ax.set_ylabel('Mean score of 100 consecutive episodes')
ax.set_title('Acrobot-v1')
```
